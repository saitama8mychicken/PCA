```python
from numpy import array
from numpy import mean
from numpy import cov
from numpy.linalg import eig
import numpy as np
class PCA():
    
    def __init__(self, n_components=None, copy=True):
        self.n_components = n_components
        self.copy = copy
        
    def fit(self, X, y=None):
        """Fit the model with X.
        Parameters
        ----------
        X : array-like, shape (n_samples, n_features)
            Training data, where n_samples is the number of samples
            and n_features is the number of features.
        y : None
            Ignored variable.
        Returns
        -------
        self : object
            Returns the instance itself.
        """
        self._fit(X)
        return self
    def fit_transform(self, X, y=None):
        """Fit the model with X and apply the dimensionality reduction on X.
        Parameters
        ----------
        X : array-like, shape (n_samples, n_features)
            Training data, where n_samples is the number of samples
            and n_features is the number of features.
        y : None
            Ignored variable.
        Returns
        -------
        X_new : array-like, shape (n_samples, n_components)
            Transformed values.
        """
        p, values = self._fit(X)
        out = pd.DataFrame(p)
        sort = self.sort_index(values)[0:self.n_components]
        return np.array(out[out.columns[sort]])
    def transform(self, X, y=None):
        return self.fit_transform(X)
    
    def _fit(self, X):
        A = pd.DataFrame(X)
        M = mean(A.T, axis=1)
        C = A - M
        V = cov(C.T)
        
        values, vectors = eig(V)
        self.components_ = vectors.T
        out_vectors = vectors
        P = vectors.T.dot(C.T)
        
        # defining explained variance be eigen vector
        self.explained_variance = sorted(values)[0:self.n_components]
        # defining explained variance ratio by eigen vector
        self.explained_variance_ratio = self.explained_variance/np.sum(self.explained_variance)
        return P.T, values
    def sort_index(self, values):
        sorted_index = []
        del_values = list(values)
        for i in range(len(del_values)):
            ind = 0
            maxi = max(del_values)
            for item in values:
                if abs(maxi-item) < 0.001:
                    sorted_index.append(ind)
                    del_values.remove(maxi)
                    break;
                else:
                    ind+=1
        return sorted_index
```
